```python
filepath = '../input/titanic/train.csv'
ttest = pd.read_csv('../input/titanic/test.csv')
tdata = pd.read_csv(filepath)
tdata.head()
```

```python
for i in [tdata,ttest]:
    i['Sex_binary']=i['Sex'].map({'male':1,'female':0})
    
tdata['Age'] = tdata['Age'].fillna(0)
ttest['Age'] = ttest['Age'].fillna(0)

tdata['Fare'] = tdata['Fare'].fillna(0)
ttest['Fare'] = ttest['Fare'].fillna(0)

#tdata['Age'] = tdata['Age'].fillna(0)
#ttest['Age'] = ttest['Age'].fillna(0)
```

```python
y = tdata.Survived
fea = ['Pclass','Sex_binary','SibSp','Parch','Fare','Age']
X = tdata[fea]
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
#tmodel = LinearRegression()
tmodel = DecisionTreeRegressor(max_depth=100,random_state = 1)
#tmodel1 = DecisionTreeRegressor(random_state = 1)
tmodel.fit(X, y)

tmodel.score(X,y)
```

```python
X_test = ttest[fea]
predictions = tmodel.predict(X_test).astype(int)
predictions


#Visualize the first 5 rows

```
